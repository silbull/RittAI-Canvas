```python
import os
import json

# print(os.getcwd())

# JSONファイルがあるファイルパス
print("READ Start")
file_path = 'syllabi.json'

# Open関数でファイルをロード
with open(file_path) as f:
    data = json.loads(f.read())
    
print("READ Finished")
```

```python
import pprint

def Search_Lecture_name(data:list, word:str) -> list:
    try:
        results = list(filter(lambda item : item["name"].find(word) >= 0, data))
    except IndexError as e:
        print("【", e, "】ヒットしませんでした．")
    
    print("【結果】", len(results), "件ヒットしました．")
    
    name_dict = {}
    for lecture in results:
        if not lecture["name"] in name_dict.keys():
            name_dict[lecture["name"]] = 1
        else:
            name_dict[lecture["name"]] += 1
    
    # pprint.pprint(name_dict)
    return results
```

```python
Search_Lecture_name(data, "アルゴリズム")
```

```python
# シラバスのJSONのリスト"l"から，"k" == "v" の条件に合うシラバスを検索し，リストとして返す．
# "p_flg"がTrueの場合は検索結果をprintする．
def Search_Lecture_JSON(l:list, k:str, v:str, p_flg:int=0) -> list:
    try:
        results = list(filter(lambda item : item[k] == v, l))
    except IndexError as e:
        print("【", e, "】ヒットしませんでした．")

    if (p_flg):
        print("【結果】", len(results), "件ヒットしました．")
        # print(results)

    return results
```

```python
search_result_01 = Search_Lecture_JSON(data, "name", "ディジタル信号処理(A,C)", p_flg=1)
print(search_result_01)
```

```python
def Print_Schedule(schedule):
    global output
    if type(schedule) == list:
        output += "【schedule】" + "\n"
        for week in schedule:
            output += str(week) + "\n" 

    elif type(schedule) == dict:
        output += "【schedule】" + "\n"
        for part, content in schedule.items():
            output += "PART" + str(part) + " " + str(content) + "\n" 
    else:
        raise ValueError("scheduleの型を確認してください．現在はlistとdictにのみ対応しています．")
```

```python
def Print_Bow(bow:list):
    global output
    output += "【bow】" + "\n" 
    # print("bow")
    for word in bow.items():
        output += str(word) + "\n" 
        # print(word)
```

```python
# シラバスの検索結果から指定したkey要素のみをprintする関数
# "selected_keys"に表示したいkeyを任意の数記載
def Print_Selected_key(syllabi_list:list,*selected_keys, ALL:bool=0 ):
    global output
    
    for syllabus in syllabi_list:
        status = []
        status.append(syllabus["name"])
        status.append(syllabus["teachers"])
        status.append(syllabus["timetable"])

        output += "【INPUT】" + str(status) + "\n"
        # print(status)

        if ALL:
            output += pprint.pformat(syllabus, sort_dicts=False)
            output += "\n\n\n"  
            continue

        elements = []
        for selected_key in selected_keys:
            if selected_key == "schedule":
                Print_Schedule(syllabus[selected_key])
                print(syllabus[selected_key])
            elif selected_key == "bow":
                Print_Bow(syllabus[selected_key])
            else:
                output += "【" + str(selected_key) + "】\n" + str(syllabus[selected_key]) + "\n"
                # print(selected_key, "\n", syllabus[selected_key])

            elements.append(syllabus[selected_key])

        output += "\n"   
        # print("")
        # print(output)
        
```
