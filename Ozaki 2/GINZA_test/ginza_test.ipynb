```python
with open(QUESTION_INPUT_PATH, "r") as f:
    question_dict = json.load(f)

pprint.pprint(question_dict, width=150)
```

```python
token_data = [
            "text",
            "lemma",
            "pos",
            "tag",
            "dep",
            "shape",  
            "is_alpha",
            "is_stop",
]

for i, value in question_dict.items():
    doc = nlp(value["English"])
    tokens_list = []
    for token in doc:
        token_data = {
            "text":     token.text,     # テキスト
            "lemma":    token.lemma_,   # レンマ
            "pos":      token.pos_,     # 品詞
            "tag":      token.tag_,     # 品詞詳細
            "dep":      token.dep_,     # 構文従属関係
            "shape":    token.shape_,   # 正書法の特徴(x:文字,d:数値)
            "is_alpha": token.is_alpha, # 文字かどうか
            "is_stop":  token.is_stop   # ストップリストの一部かどうか
        }
        tokens_list.append(token_data)
    df_tokens = pd.DataFrame(data = tokens_list , columns = token_data)
    break

df_tokens
```

```python
displacy.render(
    doc, 
    style = 'dep', 
    jupyter = True, 
    options = {
        'compact':True,
        'distance': 80
    }
)
```

```python
# hello , world の順番に並んでいるトークンのみ検出するmatcher
nlp_en = spacy.load("en_core_web_sm")
matcher = Matcher(nlp_en.vocab)

pattern = [
    [{"LOWER": "hello"}, {"IS_PUNCT": True}, {"LOWER": "world"}]
]
matcher.add("Matcher_HelloWorld", pattern, on_match=None)

doc2 = nlp_en("Hello, world! Hello. world! hello world")
matches = matcher(doc2)
for match_id, start, end in matches:
    string_id = nlp_en.vocab.strings[match_id] 
    span = doc2[start:end] 
    print(f"""match_id = {match_id} 
          string_id = {string_id}
          start = {start}
          end = {end}
          span.text = {span.text}""")
for i in range(len(doc2)):
    print(f"No.{i} {doc2[i]}")
```
